```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
import seaborn as sns
```

# N Red Balls

```python
pdf = []
t = 3
s = 2
l = range(0,11)
for N in l:
    pdf.append(math.comb(t,s)*(N/10)**s*((10-N)/10)**(t-s))
plt.plot(l,pdf)
print(l[np.argmax(pdf)])
plt.xlabel("Number of Red Balls")
plt.ylabel("Probability")
# plt.scatter(l[np.argmax(pdf)], max(pdf) , color='r');
plt.axvline(x=l[np.argmax(pdf)],color='r')
plt.title("Q.2 N Red Balls")
```

```python
l[np.argmax(pdf)], max(pdf)
```

```python
pdf = []
n = 30
l = range(0,31)
for r in l:
    pdf.append(math.comb(n,r)*(0.1)**r*(0.9)**(n-r))
plt.figure(figsize=(10,5))
plt.plot(l,pdf)
print(l[np.argmax(pdf)])
plt.xlabel("Number of Clusters with Dominant Galaxies")
plt.ylabel("Probability")
plt.axvline(x=l[np.argmax(pdf)],color='r')
plt.title("Q.3 Dominant Central Galaxies")
```

# Coin Toss

```python
def lead_calc(N):
    A = [0]
    B = [0]
    for i in range(N):
        c = random.randint(0,1)
        if c>0:
            # when A wins
            A.append(A[-1]+1)
            B.append(B[-1]-1)
        else:
            # When B wins
            A.append(A[-1]-1)
            B.append(B[-1]+1)
    a=0
    b=0
    lead_changes_at = []
    for i in range(1, len(A)):
        if A[i]>B[i]:
            if A[i-1]>B[i-1]:
                pass
            else:
                a=a+1
                lead_changes_at.append(i)
        elif A[i]<B[i]:
            if A[i-1]<B[i-1]:
                pass
            else:
                b=b+1
                lead_changes_at.append(i)
        else:
            pass
    return a+b, lead_changes_at[-1], abs(A[-1]-B[-1]), A, B 

leads = []
lead_changed_at = [] #last lead changed at
leads_by = []
for i in range(10000):
    a,b,c,A,B = lead_calc(1000)
    leads.append(a)
    lead_changed_at.append(b)
    leads_by.append(c)
```

```python
# Plot of Toss Number vs Wins/Losses
plt.plot(A)
plt.plot(B)
plt.xlabel("Toss Number")
plt.ylabel("No. of Wins/Losses")
plt.legend(['Player A','Player B'])
plt.show()
```

```python
# Average Number of Lead Changes
sns.boxplot(x=leads)
plt.xlabel("Number of times leads changed")
np.mean(leads)
```

```python
# Leads changed at
sns.histplot(lead_changed_at)
plt.xlabel("Toss Number")
plt.title("Number of times lead changed at given toss number")
```

```python
# Average number of leads by
sns.boxplot(x=leads_by)
plt.xlabel("Leads By")
np.mean(leads_by)
```
